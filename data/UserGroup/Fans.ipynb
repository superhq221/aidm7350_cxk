```python
%matplotlib inline
import jieba
import pandas as pd
import re
import matplotlib.pyplot as plt 
import matplotlib
import seaborn as sns
from wordcloud import WordCloud, ImageColorGenerator
```

```python
fans_data1 = pd.read_csv('/Users/nancy/Desktop/Attempt/data/cxk1.csv')
```

```python
fans_data1
```

```python
userID = fans_data1['Usename']
IDWhole = "".join(userID.tolist())
ID_Segment = jieba.lcut(IDWhole, cut_all = False, HMM = True)
```

```python
stopWords = []
with open('/Users/nancy/Desktop/Attempt/data/stopwords-master/百度停用词表.txt','r', encoding = 'utf-8') as stopWordTxt:
    for word in stopWordTxt:
        stopWords.append(re.sub(r'[\n]+','',word))
stopWords.append('_')
stopWords.append('-')
stopWords.append('·')
```

```python
for seg in ID_Segment:
    if seg in stopWords:
        ID_Segment.remove(seg)
```

```python
countD = {}
for user in ID_Segment:
    if user in countD:
        countD[user] += 1
    else:
        countD[user] = 1
```

```python
dfTest = pd.DataFrame()
dfTest['seg'] = countD.keys()
dfTest['count'] = countD.values()
```

```python
dfTest.sort_values(by = 'count', ascending= False).head(30)
```

```python
dfTest.to_csv('fansID1.csv')
```
