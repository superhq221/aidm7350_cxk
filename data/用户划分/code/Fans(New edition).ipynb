```python
userID = fans_data['Usename']
IDWhole = "".join(userID.tolist())
ID_Segment = jieba.lcut(IDWhole, cut_all = False, HMM = True)
```

```python
specialWord = ['蔡徐坤','蔡小葵','小葵']
for word in specialWord:
    jieba.add_word(word)
```

```python
stopWords = []
with open('/Users/nancy/Desktop/Attempt/data/stopwords-master/百度停用词表.txt','r', encoding = 'utf-8') as stopWordTxt:
    for word in stopWordTxt:
        stopWords.append(re.sub(r'[\n]+','',word))
stopWords.append('_')
stopWords.append('-')
stopWords.append('·')
stopWords.append('丶')
stopWords.append('__')
stopWords.append('_')
stopWords.append('_')
```

```python
for seg in ID_Segment:
    if seg in stopWords:
        ID_Segment.remove(seg)
```

```python
countD = {}
for user in ID_Segment:
    if user in countD:
        countD[user] += 1
    else:
        countD[user] = 1
```
